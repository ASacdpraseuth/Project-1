```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import requests
import time
import json
```

```python
## https://github.com/M-Media-Group/Covid-19-API
# so far only using cases.

base_url = "https://covid-api.mmediagroup.fr/v1"
case_url = "/cases"
history_url = "/history"
vaccine_url = "/vaccines"
```

```python
response = requests.get(base_url + case_url).json()

data = json.dumps(response, indent=4, sort_keys=True)
print(data)
```

```python
country_dict = {
    "location": [],
    "continent": [],
    "country": [],
    "abbreviation": [],
    "lat": [],
    "long": [],
    "sq_km_area": [],
    "population": [],
    "confirmed": [],
    "recovered": [],
    "deaths": [],
    "updated": []
}

split_countries = []

for index, country in enumerate(response):
    
    try:
        print(f"Appending index {index} of {len(response)}: {country}.")
        country_dict["location"].append(response[country]["All"]["location"])
    except(KeyError) as e:
        country_dict["location"].append("Incomplete")
    try:
        country_dict["continent"].append(response[country]["All"]["continent"])
    except(KeyError) as e:
        country_dict["continent"].append("Incomplete")
    try:
        country_dict["country"].append(country)
    except(KeyError) as e:
        country_dict["country"].append("Incomplete")
    try:
        country_dict["abbreviation"].append(response[country]["All"]["abbreviation"])
    except(KeyError) as e:
        country_dict["abbreviation"].append("Incomplete")
    try:
        country_dict["lat"].append(response[country]["All"]["lat"])
    except(KeyError) as e:
        country_dict["lat"].append("Incomplete")
    try:
        country_dict["long"].append(response[country]["All"]["long"])
    except(KeyError) as e:
        country_dict["long"].append("Incomplete")
    try:
        country_dict["sq_km_area"].append(response[country]["All"]["sq_km_area"])
    except(KeyError) as e:
        country_dict["sq_km_area"].append("Incomplete")
    try:
        country_dict["population"].append(response[country]["All"]["population"])
    except(KeyError) as e:
        country_dict["population"].append("Incomplete")
    try:
        country_dict["confirmed"].append(response[country]["All"]["confirmed"])
    except(KeyError) as e:
        country_dict["confirmed"].append("Incomplete")
    try:
        country_dict["recovered"].append(response[country]["All"]["recovered"])
    except(KeyError) as e:
        country_dict["recovered"].append("Incomplete")
    try:
        country_dict["deaths"].append(response[country]["All"]["deaths"])
    except(KeyError) as e:
        country_dict["deaths"].append("Incomplete")
    try:
        country_dict["updated"].append(response[country]["All"]["updated"])
    except(KeyError) as e:
        country_dict["updated"].append("Incomplete")
 
```

```python
raw_data = pd.DataFrame(country_dict)
raw_data
```
