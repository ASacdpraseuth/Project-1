```python
import requests
import json
import pandas as pd
```

```python
owid_locations = pd.read_csv("csv_files/owid_covid19_data.csv")
owid_locations.head()
```

```python
len(owid_locations)
```

```python
base_url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.json"
```

```python
response = requests.get(base_url).json()

#results = response[i]



print(json.dumps(response[0], indent=4, sort_keys=True))
```

```python
json_dict = {
    "country": [],
    "iso_code": [],
    "date": [],
    "daily_vaccinations": [],
    "daily_vaccinations_per_million": []
}

for i in range(0, len(owid_locations)):
    results = response[i]

    for index, data in enumerate(results["data"]):
        json_dict["country"].append(results["country"])
        json_dict["iso_code"].append(results["iso_code"])
            
        json_dict["date"].append(results["data"][index]["date"])
            
        try:
            json_dict["daily_vaccinations"].append(results["data"][index]["daily_vaccinations"])
        except(KeyError):
            json_dict["daily_vaccinations"].append("No Data")
                
        try:    
            json_dict["daily_vaccinations_per_million"].append(results["data"][index]["daily_vaccinations_per_million"])
        except(KeyError):
            json_dict["daily_vaccinations_per_million"].append("No Data")
```

```python
raw_data_df = pd.DataFrame(json_dict)

print(len(raw_data_df))

raw_data_df
```
